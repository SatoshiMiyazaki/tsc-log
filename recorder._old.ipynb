### データベースへの登録

#### dbNameというデータベースにdevName_typeというtableを作成する。devNameにハイフン-が入る場合は、アンダースコア_に置換される。

```python
import pandas as pd
df = pd.read_csv('dump-tsc-loghsc.csv', low_memory=False)
df.info()
```

```python
import datetime
from telDav import RegisterData

import threading

threads = []

for devName in df['DevName'].drop_duplicates():
    if type(devName) != str:
        continue
    if 'PMA' in devName:
        continue
#    if 'MLP2' in devName:
#        continue # because it is too large
    print(devName, end=' ')

    dataType = 'E'

    registerData = RegisterData.RegisterData(startDateTime = datetime.datetime(year=2021, month=1, day=1), 
                                         durationInDay = 365*4, 
                                         dataType = dataType, 
                                         devName = devName, 
                                         dbName = f'db/{devName.replace("-", "_")}_{dataType}.db',
                                         isVerbose = False
                                        )
    if registerData.numberOfNames > 0:
        threads.append(threading.Thread(target=registerData.run))

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()
    
```
